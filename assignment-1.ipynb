# 1). Program to find the number , divisible by 7 but not multiple of 5 between             2000 and 3200 (included both)

```python
def find_num():
    l=[]
    for no in range(2000,3201):
        if(no%7==0 and (str(no)[-1]!='0' and str(no)[-1]!='5')):
            l.append(no)
    print(','.join(str(x) for x in l))

obj=find_num()
obj  
```

# 3). Program to find the volume of a sphere having diameter 12 cm.

```python
from math import pi
def find_vol(dia=12):
    r=dia/2
    print('Volume of sphere: ',4/3*pi*r**3)
find_vol()
```
